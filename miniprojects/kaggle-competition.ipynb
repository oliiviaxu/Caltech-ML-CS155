## **Data Processing**


```python
#handle NAN data
train_df.isna().sum()
train_df = train_df.fillna(0)
test_df = test_df.fillna(0)

train_data = pd.DataFrame()
test_data = pd.DataFrame()
```

```python
def preprocess_data(data, orig_df):
    # One hot encoding for gender
    data['female'] = [1 if x == 'Female' else 0 for x in orig_df['gender']]
    data['male'] = [1 if x == 'Male' else 0 for x in orig_df['gender']]

    # Senior citizen - leave alone
    data['senior_citizen'] = orig_df['SeniorCitizen']

    # Partner and dependents (yes-no to 1-0 mapping)
    data['partner'] = [1 if x == 'Yes' else 0 for x in orig_df['Partner']]
    data['dependents'] = [1 if x == 'Yes' else 0 for x in orig_df['Dependents']]

    # Tenure ranges from 1-72, leave it as is
    data['tenure'] = orig_df['tenure']

    # Phone service, multiple lines (yes-no to 1-0 mapping)
    data['phone_service'] = [1 if x == 'Yes' else 0 for x in orig_df['PhoneService']]
    data['multiple_lines'] = [1 if x == 'Yes' else 0 for x in orig_df['MultipleLines']]

    # Splitting Internet Service into its 3 options - DSL, Fiber Optic, and None
    data['dsl'] = [1 if x == 'DSL' else 0 for x in orig_df['InternetService']]
    data['fibre_optic'] = [1 if x == 'Fiber optic' else 0 for x in orig_df['InternetService']]

    # No_service value is the same for internet service, online security, online backup,
    # device protection, tech support, and streaming TV. Goes for all columns, so
    # we only need 1 column to indicate that
    data['no_internet_service'] = [1 if x == 'No' else 0 for x in orig_df['InternetService']]

    #model performed better without these columns
    # Encode "Online Security" with one-hot encoding
    # data['online_security_yes'] = [1 if x == 'Yes' else 0 for x in orig_df['OnlineSecurity']]
    # data['online_security_no'] = [1 if x == 'No' else 0 for x in orig_df['OnlineSecurity']]

    # # Encode "Online Backup" with one-hot encoding
    # data['online_backup_yes'] = [1 if x == 'Yes' else 0 for x in orig_df['OnlineBackup']]
    # data['online_backup_no'] = [1 if x == 'No' else 0 for x in orig_df['OnlineBackup']]

    # # Encode "Device Protection" with one-hot encoding
    # data['device_protection_yes'] = [1 if x == 'Yes' else 0 for x in orig_df['DeviceProtection']]
    # data['device_protection_no'] = [1 if x == 'No' else 0 for x in orig_df['DeviceProtection']]

    # # Encode "Tech Support" with one-hot encoding
    # data['tech_support_yes'] = [1 if x == 'Yes' else 0 for x in orig_df['TechSupport']]
    # data['tech_support_no'] = [1 if x == 'No' else 0 for x in orig_df['TechSupport']]

    # Encode "Streaming TV" with one-hot encoding
    data['streaming_tv_yes'] = [1 if x == 'Yes' else 0 for x in orig_df['StreamingTV']]
    data['streaming_tv_no'] = [1 if x == 'No' else 0 for x in orig_df['StreamingTV']]

    # Encode "Streaming Movies" with one-hot encoding
    data['streaming_movies_yes'] = [1 if x == 'Yes' else 0 for x in orig_df['StreamingMovies']]
    data['streaming_movies_no'] = [1 if x == 'No' else 0 for x in orig_df['StreamingMovies']]

    # One-hot encoding for contract
    data['monthly_contract'] = [1 if x == 'Month-to-month' else 0 for x in orig_df['Contract']]
    data['one_year_contract'] = [1 if x == 'One year' else 0 for x in orig_df['Contract']]
    data['two_year_contract'] = [1 if x == 'Two year' else 0 for x in orig_df['Contract']]

    # Paperless billing (yes-no to 1-0 mapping)
    data['paperless_billing'] = [1 if x == 'Yes' else 0 for x in orig_df['PaperlessBilling']]

    #Dropping payment method column

    # Monthly charges and total charges
    data['monthly_charges'] = orig_df['MonthlyCharges']
    data['total_charges'] = orig_df['TotalCharges']

    return data
```

```python
train_data = preprocess_data(train_data, train_df)
train_labels = pd.DataFrame()
train_labels['discontinued'] = [1 if x == 'Yes' else 0 for x in train_df['Discontinued']]
display(train_data)
display(train_labels)
```

```python
test_data = preprocess_data(test_data, test_df)
display(test_data)
```

# **Models**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from scipy.stats import randint
from sklearn.model_selection import GridSearchCV
```

**AdaBoost**

We will implement an AdaBoost Hyperparameters Grid Search. Guide followed: https://www.kaggle.com/code/bachnguyentfk/adaboost-hyperparameters-grid-search This was our best performing model.

```python
model = AdaBoostClassifier()
#Grid search implementation for hyperparamater tuning
grid = dict()
grid['n_estimators'] = [10, 50, 100, 500]
grid['learning_rate'] = [0.0001, 0.001, 0.01, 0.1, 1.0]

# define the evaluation procedure
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy')

grid_result = grid_search.fit(train_data, train_labels)

# summarize the best score and configuration
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score

best_n_estimators = grid_result.best_params_['n_estimators']
best_learning_rate = grid_result.best_params_['learning_rate']

adaboost_model = AdaBoostClassifier(n_estimators=best_n_estimators, learning_rate=best_learning_rate)

cv_scores = cross_val_score(adaboost_model, train_data, train_labels, cv=10, scoring='roc_auc')

for fold, score in enumerate(cv_scores, start=1):
    print(f"Fold {fold}: ROC AUC = {score:.4f}")

adaboost_model.fit(train_data, train_labels)
```

```python
submission = pd.DataFrame()
submission['ID'] = test_df['customerID']
submission['TARGET'] = adaboost_model.predict_proba(test_data)[:, 1]

display(submission)
```

```python
submission.to_csv('ada_submission.csv', index=False)
```

**Gradient boosting (XG Boost)**

```python
import xgboost as xgb
from xgboost import cv
from xgboost import XGBClassifier
```

We are doing k-fold Cross Validation with XGBoost, following the guide:
https://gist.github.com/pb111/cc341409081dffa5e9eaf60d79562a03. Before doing so we will do a Randomized search to find a good combination of hyperparamaters.

```python
param_grid = {
    'objective': ['binary:logistic'],
    'colsample_bytree': [0.3, 0.5, 0.7, 1.0],
    'subsample': [0.6, 0.8, 1.0],
    'learning_rate': [0.001, 0.05, 0.1, 0.2],
    'max_depth': [3, 4, 5, 6, 7],
    'alpha': [1, 5, 10],
    'gamma': [0.5, 1, 1.5, 2, 5],
    'n_estimators': [50, 100, 200, 500, 1000],
}

xgb_clf = XGBClassifier()

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

random_search = RandomizedSearchCV(xgb_clf, param_distributions=param_grid, n_iter=10,
                                   scoring='accuracy', cv=cv, verbose=1, random_state=42, n_jobs=-1)

random_search.fit(train_data, train_labels)

print(random_search.cv_results_)
print(random_search.best_params_)
```

```python
# Define a parameter grid for the grid search
param_grid = {
    'objective': ['binary:logistic'],
    'colsample_bytree': [0.3, 1],
    'subsample': [0.8, 1.0],
    'learning_rate': [0.1, 0.2],
    'max_depth': [5],
    'alpha': [1, 5],
    'gamma': [1.5],
    'n_estimators': [100, 200],
}

xgb_clf = XGBClassifier()

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

# Create a grid search with cross-validation
grid_search = GridSearchCV(xgb_clf, param_grid=param_grid, scoring='accuracy', cv=cv, verbose=1, n_jobs=-1)

grid_search.fit(train_data, train_labels)

# Print the results
print("Grid Search Best Parameters:", grid_search.best_params_)
print("Grid Search Best Score (Accuracy): {:.4f}".format(grid_search.best_score_))
```

k-fold cross validation using XGBoost

```python
data_dmatrix = xgb.DMatrix(data=train_data,label=train_labels)

params = {"objective":"binary:logistic",'learning_rate': 0.2,
                'max_depth': 5, 'alpha': 5, 'n_estimators':200}

xgb_clf = XGBClassifier(n_estimators=200, learning_rate=0.2, max_depth=5, alpha=5)
xgb_clf.fit(train_data, train_labels)
print(xgb_clf)

#k fold cross validation
xgb_cv = cv(dtrain=data_dmatrix, params=params, nfold=3,
                    num_boost_round=100, metrics="auc", as_pandas=True)
xgb_cv.head()
```

```python
#submission template
submission = pd.DataFrame()
submission['ID'] = test_df['customerID']
submission['TARGET'] = xgb_clf.predict_proba(test_data)[:, 1]

display(submission)
```

GridSearchCV

```python
param_grid = {
    'n_estimators': [100, 500],
    'max_depth': [None, 10],
    'min_samples_leaf': [15, 30],
    'max_features': ['sqrt', 'log2'],
    'random_state': [25, 50]
}

rf_classifier = RandomForestClassifier()

# Grid Search
# Parameter options
params = GridSearchCV(
    rf_classifier,
    param_grid=param_grid,
    cv=5,  # of cross-validation folds
    scoring='roc_auc',
    return_train_score=True,
    n_jobs=-1,
)

params.fit(X_train, y_train)

best_params = params.best_params_
print("Best Hyperparameters:", best_params)

best_rf_model = params.best_estimator_

# y_pred_prob = best_rf_model.predict_proba(X_test)[:, 1]
```

```python
best_rf_model = params.best_estimator_

y_pred_binary = best_rf_model.predict(X_train)

accuracy = accuracy_score(y_train, y_pred_binary)
print("Accuracy:", accuracy)
```
